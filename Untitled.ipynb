# Data Cleaning

```python
# Create a copy of the data, but with only the columns I'll be working with

df = data.drop(['id', 'date', 'sqft_lot', 'sqft_above', 'view', 'yr_built','yr_renovated', 'address', 'lat', 'long'], axis=1)
```

```python
# Dropping houses with 0 bedrooms or 0 bathrooms

df = df[df['bedrooms'] != 0]
df = df[df['bathrooms'] != 0]
```

```python
# Working with bedrooms 2,3,4

# df = df[(df['bedrooms'] == 2) | (df['bedrooms'] == 3) | (df['bedrooms'] == 4)]
```

```python
# Dropping NaN values

df = df.dropna()
```

```python
# Make features that are No and Yes to binary 0s and 1s

df.waterfront = df.waterfront.map(dict(YES=1, NO=0))
df.greenbelt = df.greenbelt.map(dict(YES=1, NO=0))
df.nuisance = df.nuisance.map(dict(YES=1, NO=0))
```

```python
# Getting rid of the Restricted section for both of them since there are only 3 & 6 values 

df.sewer_system = df.sewer_system.replace(to_replace='PRIVATE RESTRICTED', value='PRIVATE')
df.sewer_system = df.sewer_system.replace(to_replace='PUBLIC RESTRICTED', value='PUBLIC')
```

```python
# Changing sqft from basement, garage, and patio into binary to see if they have it or they don't have it

df['sqft_basement'] = df['sqft_basement'].apply(lambda x: 1 if x > 0 else 0)
df['sqft_garage'] = df['sqft_garage'].apply(lambda x: 1 if x > 0 else 0)
df['sqft_patio'] = df['sqft_patio'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Renaming column names

df.rename(columns = {'sqft_basement':'basement', 'sqft_garage':'garage', 'sqft_patio':'patio'}, inplace = True)
```
