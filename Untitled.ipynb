```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
import statsmodels.api as sm
import seaborn as sns
from scipy import stats

%matplotlib inline
```

```python
# Collect entire data 

data = pd.read_csv('data/kc_house_data.csv')
```

```python
# Checkout all data columns & spot Null values

data.info()
```

# Data Cleaning

```python
# Create a copy of the data, but with only the columns I'll be working with

df = data.drop(['id', 'date', 'sqft_lot', 'sqft_above', 'view', 'yr_built','yr_renovated', 'address', 'lat', 'long'], axis=1)
```

```python
# Dropping houses with 0 bedrooms or 0 bathrooms

df = df[df['bedrooms'] != 0]
df = df[df['bathrooms'] != 0]
```

```python
# Working with bedrooms 2,3,4

# df = df[(df['bedrooms'] == 2) | (df['bedrooms'] == 3) | (df['bedrooms'] == 4)]
```

```python
# Dropping NaN values

df = df.dropna()
```

```python
# Make features that are No and Yes to binary 0s and 1s

df.waterfront = df.waterfront.map(dict(YES=1, NO=0))
df.greenbelt = df.greenbelt.map(dict(YES=1, NO=0))
df.nuisance = df.nuisance.map(dict(YES=1, NO=0))
```

```python
# Getting rid of the Restricted section for both of them since there are only 3 & 6 values 

df.sewer_system = df.sewer_system.replace(to_replace='PRIVATE RESTRICTED', value='PRIVATE')
df.sewer_system = df.sewer_system.replace(to_replace='PUBLIC RESTRICTED', value='PUBLIC')
```

```python
# Changing sqft from basement, garage, and patio into binary to see if they have it or they don't have it

df['sqft_basement'] = df['sqft_basement'].apply(lambda x: 1 if x > 0 else 0)
df['sqft_garage'] = df['sqft_garage'].apply(lambda x: 1 if x > 0 else 0)
df['sqft_patio'] = df['sqft_patio'].apply(lambda x: 1 if x > 0 else 0)
```

```python
# Renaming column names

df.rename(columns = {'sqft_basement':'basement', 'sqft_garage':'garage', 'sqft_patio':'patio'}, inplace = True)
```

```python
# Manually deleting the first column from heat_source

# dummy_heat_source.drop('Gas', axis=1)
```

```python
# getting rid of the outliers on top & bottom .025%

df_clean = df.copy()

outliers = ['sqft_living','price']

for col in outliers:
    df_clean = df_clean[(df_clean[col] > np.percentile(df_clean[col], .25)) & (df_clean[col] < np.percentile(df_clean[col], 99.75))]
```

# Graphs

```python
def format_number(data_value,indx):
    if data_value >= 1000000:
        formatter = '{:1.1f}M'.format(data_value * 0.000001)
    else:
        formatter = '{:1.1f}K'.format(data_value * 0.001)
    return formatter
```

```python
sns.set_theme(style='whitegrid')
sns.set(rc={"figure.dpi":100, 'savefig.dpi':100})
```

```python
fig, ax = plt.subplots()
sns.barplot(x='waterfront', y='price', data=ddf)

ax.set_title('Average Price Difference No Waterfront VS Waterfront')
plt.xlabel("Waterfront")
plt.ylabel("Price")

ax.yaxis.set_major_formatter(format_number)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot(x='garage', y='price', data=ddf)

ax.set_title('Average Price Difference No Garage VS Garage')
plt.xlabel("Garage")
plt.ylabel("Price")

ax.yaxis.set_major_formatter(format_number)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.barplot(x='patio', y='price', data=ddf)

ax.set_title('Average Price Difference No Patio VS Patio')
plt.xlabel("Patio")
plt.ylabel("Price")

ax.yaxis.set_major_formatter(format_number)

plt.tight_layout()
plt.show()
```
